```python
import cv2
import numpy as np
import pytesseract
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
def remove_background(img_path):
    image = cv2.imread(img_path)
    img_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    _,img_thresh = cv2.threshold(img_gray,10,255,cv2.THRESH_BINARY) # set a threshold to assigning 0 using cv2.THRESH_BINARY Method
    denoise = cv2.medianBlur(img_thresh,7)  #reduces the noise
    output = cv2.GaussianBlur(denoise,(5,5),0) #Blurs an image using GaussianBlur
    return output
```

```python
img_path = './Computer Vision/Task-1b.png'
```

```python
output = remove_background(img_path)
```

```python
plt.figure(figsize=(20,20))
plt.imshow(output)
```

```python
config = ('-l eng --oem 1 --psm 12')
text = pytesseract.image_to_string(output, config=config)
```
